```python
import warnings
warnings.filterwarnings("ignore", category = FutureWarning)

import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
src_dir = './SINGERS/'
song_samples = 6600000
singers = {"Arjit_Singh":0,"Armaan":1,"Atif Aslam":2,"Sonu Nigam":3}
```

```python
window =.1
overlap =.5
X = 660000
xshape = X
print(xshape)
chunk = int(xshape*window)
print(chunk)
offset = int(chunk*(1.-overlap))
print(offset)
hop_length = 512 
n_fft = 1024
x= []
y=[]

  
```

```python
for filenames in (os.listdir(src_dir)):
        
            print("\nProcessing: {}".format(filenames))
            for f in os.listdir("{}\\{}".format(src_dir,filenames)) :
            
                file_path ="{}\\{}\\{}".format(src_dir,filenames,f)
                signal, sr = librosa.load(file_path, sr=22000)
             
                print(file_path,":",signal) 
                
                for t in range(0,len(signal),660000):
                    if(t+660000 > len(signal)-1):
                        break
                        
                    else :   
                
                        print(t,t+660000)
                        X = signal[t:t+660000]
                        print("Reading file: {}".format(file_path))
                
                 # Convert to dataset of spectograms/melspectograms
                        g = singers[filenames]
                        
                        for i in range(0, xshape-chunk + offset, offset):
                           ## splitting song sample into parts 
                             p = X[i:i+chunk]
                             print(i,i+chunk)
                             k=librosa.feature.melspectrogram(p, n_fft = n_fft,hop_length = hop_length)[:,:,np.newaxis]
                             ##converting amplitude signals to mel spectograms for each part
                             print(k.shape)
                       
                             x.append(k)
                             y.append(g)
                             print("genre no :" ,g)
                             print("no of mel spectograms till now" ,len(x))

                        
```

```python
for i in range(4):
    print("No of mel Spectograms for Music Artist",list(singers.keys())[i],"-->",y.count(i))
```

```python
np.save('./data3/x_singers_npy.npy', np.array(x))
np.save('./data3/y_singers_npy.npy', np.array(y))
```
